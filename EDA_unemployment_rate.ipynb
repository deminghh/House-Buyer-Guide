```python
import requests
import pandas as pd
import json
from cred import credentials
from request_api import get_genesis_api
```

```python
df_unemployment_rate = pd.read_csv('data/13211-0007.csv', delimiter= ';', header= [0,1], skiprows= 4, skipfooter=4, encoding= 'latin1', engine='python')
df_unemployment_rate
```

```python
df_unemployment_rate.columns = df_unemployment_rate.columns.map('_'.join)
df_unemployment_rate
```

```python
df_unemployment_rate.rename(columns={'Unnamed: 0_level_0_Unnamed: 0_level_1': 'bundesland', 'Unnamed: 1_level_0_Unnamed: 1_level_1': 'year'}, inplace=True)
df_unemployment_rate
```

```python
df_unemployment_rate.columns = df_unemployment_rate.columns.str.replace(' ', '_')
df_unemployment_rate
```

```python
df_unemployment_rate.drop(columns=['Arbeitslose_Anzahl', 'Arbeitslosenquote_d._abhängigen_ziv._Erwerbspers._Prozent', 'Gemeldete_Arbeitsstellen_Anzahl'], inplace=True)
df_unemployment_rate
```

```python
df_unemployment_rate.rename(columns={'Arbeitslosenquote_aller_zivilen_Erwerbspersonen_Prozent': 'unemployment_rate'}, inplace=True)
df_unemployment_rate
```

```python
df_unemployment_rate.unemployment_rate = df_unemployment_rate.unemployment_rate.str.replace(',', '.')
df_unemployment_rate
```

```python
cols = df_unemployment_rate.columns.drop(['bundesland', 'year'])

df_unemployment_rate[cols] = df_unemployment_rate[cols].apply(pd.to_numeric, errors='coerce')
df_unemployment_rate
```

```python
df_unemployment_rate.to_csv('data/cleaned_unemployment_rate.csv')
```
