```python
# function of changing th name of the counties according to the year 2020 
def change_str(string):
    """
    >>> change_str('Steig, Kreis')
    'LK Steig'
    >>> change_str('Apple, Landkreis')
    'LK Apple'
    >>> change_str(' Ulm, Stadtkreis')
    'SK Ulm'
    """
    exceptional_cases = {
        'Städteregion Aachen (einschl. Stadt Aachen)': 'Städteregion Aachen',
        'Aachen, krfr. Stadt': 'Städteregion Aachen',
        'Baden-Baden, Stadtkreis': 'SK Baden-Baden',
        'Chemnitz, Stadt': 'SK Chemnitz',
        'Dresden, Stadt': 'SK Dresden',
        'Freiburg im Breisgau, Stadtkreis': 'SK Freiburg im Breisgau',
        'Hannover, Stat. Region': 'Region Hannover',
        'Heidelberg, Stadtkreis': 'SK Heidelberg',
        'Heilbronn, Universitätsstadt, Stadtkreis': 'SK Heilbronn',
        'Karlsruhe, Stadtkreis': 'SK Karlsruhe',
        'Landkreis Ludwigslust-Parchim': 'LK Ludwigslust-Parchim',
        'Landkreis Nordwestmecklenburg': 'LK Nordwestmecklenburg',
        'Leipzig, Stadt': 'SK Leipzig',
        'Region Hannover, Landkreis': 'Region Hannover',
        'Mannheim, Stadtkreis': 'SK Mannheim',
        'München, Landeshauptstadt': 'SK München',
        'Pforzheim, Stadtkreis': 'SK Pforzheim',
        'Saarbrücken, Regionalverband': 'Regionalverband Saarbrücken',
        'Schwerin, Landeshauptstadt, kreisfreie Stadtk': 'SK Schwerin',
        'Kiel, Landeshauptstadt, kreisfreie Stadt': 'SK Kiel',
        'Lübeck, Hansestadt, kreisfreie Stadt': 'SK Lübeck',
        'Rostock, Hanse- und Universitätsstadt, kreisfreie Stadt': 'SK Rostock',
        'Wiesbaden, Landeshauptstadt, kreisfreie Stadt': 'SK Wiesbaden',
        'Stuttgart, Landeshauptstadt, Stadtkreis': 'SK Stuttgart',
        'Ulm, Stadtkreis': 'SK Ulm',
        'Regensburg': 'SK Regensburg',
    }
    trimmed = string.strip()
    if ' ' not in trimmed and '-' not in trimmed and 'kreis' not in trimmed:
        return 'SK' + ' ' + trimmed
    elif trimmed in exceptional_cases:
        return exceptional_cases[trimmed]
    elif trimmed[-len(', Landkreis'):] == ', Landkreis':
        return 'LK ' + trimmed[:-len(', Landkreis')]
    elif trimmed[-len(', kreisfreie Stadt'):] == ', kreisfreie Stadt':
        return 'SK ' + trimmed[:-len(', kreisfreie Stadt')]
    elif trimmed[-7:] == ', Kreis':
        return 'LK ' + trimmed[:-7]
    else:
        return trimmed

doctest.testmod()
```
