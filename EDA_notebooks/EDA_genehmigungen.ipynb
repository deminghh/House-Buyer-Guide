```python
import requests
import pandas as pd
import json
from cred import credentials
from request_api import get_genesis_api
```

```python
df_genehmigung = pd.read_csv('data/31111-0003.csv', delimiter=';', skiprows=6, skipfooter=4, header=[1,2], encoding = 'latin1')
df_genehmigung
```

```python
df_genehmigung.columns = df_genehmigung.columns.map('_'.join)
df_genehmigung
```

```python
df_genehmigung.columns = df_genehmigung.columns.str.replace(' ', '_')
df_genehmigung
```

```python
df_genehmigung.rename(columns={'Unnamed:_0_level_0_Unnamed:_0_level_1': 'bundesland', 'Unnamed:_1_level_0_Unnamed:_1_level_1': 'building_type', 'Unnamed:_2_level_0_Unnamed:_2_level_1': 'year'}, inplace=True)
df_genehmigung
```

```python
df_genehmigung.drop(columns=['Gebäude_Anzahl', 'Rauminhalt_1000_cbm', 'Baumaßnahmen_Anzahl'], inplace=True)
df_genehmigung
```

```python
df_genehmigung.info()
```

```python
cols = df_genehmigung.columns.drop(['bundesland', 'building_type', 'year'])

df_genehmigung[cols] = df_genehmigung[cols].apply(pd.to_numeric, errors='coerce')
df_genehmigung
```

```python
df_genehmigung.info()
```

```python
cols = df_genehmigung.columns.drop(['bundesland', 'building_type', 'year'])
cols
```

```python
df_genehmigung[cols] = df_genehmigung[cols].fillna(-1).astype(int)
df_genehmigung
```

```python
df_genehmigung_wohngebäude_nichtwohngebäude = df_genehmigung.query('building_type == "Wohn- und Nichtwohngebäude"')
```

```python
df_genehmigung_nichtwohngebäude_wohngebäude = df_genehmigung.query('building_type == "Nichtwohngebäude" or building_type == "Wohngebäude"')
df_genehmigung_nichtwohngebäude_wohngebäude
```

```python
df_genehmigung_nichtwohngebäude_wohngebäude.columns
```

```python
df_genehmigung_nichtwohngebäude_wohngebäude.drop(columns=['Nutzfläche_1000_qm', 'Wohnungen_Anzahl','Sonstige_Wohneinheiten_Anzahl','Wohnfläche_der_sonstigen_Wohneinheiten_1000_qm', 'Wohnräume_Anzahl'], inplace=True)
      
```

```python
df_genehmigung_nichtwohngebäude_wohngebäude
```

```python
df_genehmigung_wohngebäude = df_genehmigung_nichtwohngebäude_wohngebäude.query('building_type == "Wohngebäude"')
df_genehmigung_wohngebäude
```

```python
df_genehmigung_nichtwohngebäude = df_genehmigung_nichtwohngebäude_wohngebäude.query('building_type == "Nichtwohngebäude"')
df_genehmigung_nichtwohngebäude
```

```python
df_genehmigung_merged = df_genehmigung_wohngebäude.merge(df_genehmigung_nichtwohngebäude, on=['bundesland', 'year'])
df_genehmigung_merged
```

```python
df_genehmigung_merged.columns = df_genehmigung_merged.columns.str.replace('x', 'Wohngebäude')
```

```python
df_genehmigung_merged.columns = df_genehmigung_merged.columns.str.replace('y', 'Nichtwohngebäude')
```

```python
df_genehmigung_merged.drop(columns=['building_tNichtwohngebäudepe_Wohngebäude', 'building_tNichtwohngebäudepe_Nichtwohngebäude'], inplace=True)
df_genehmigung_merged
```

```python
df_genehmigung_merged.rename(columns={'Nichtwohngebäudeear':'year'}, inplace=True)
```
