```python
population.drop(columns='Unnamed: 0', inplace=True)
population
```

```python
result = population.groupby(['year', 'state']).age.mean().reset_index()
result
```

```python
result2 = population.groupby(['year', 'state']).total_population.sum().reset_index()
result2
```

```python
group_population = pd.merge(result, result2, on=['year', 'state'])
group_population.rename(columns={'age': 'avg_age'}, inplace=True)
group_population
```

```python
group_population.query('year == 2020')
```

```python
group_population.to_csv('data/grouped_population.csv')
```

```python
gb_total = population.groupby(['year', 'state']).sum().reset_index()
gb_total
```

```python
population['weights'] = population['age'] * population['total_population']
population
```

```python
gb_population = population.groupby(['year','state']).sum().reset_index()
gb_population
```

```python
gb_population['avg'] = gb_population.weights / gb_total.total_population
gb_population
```

```python
gb_population.rename(columns={'avg': 'age_avg'}, inplace=True)
gb_population
```

```python
gb_population.query('year == 2020')
```
