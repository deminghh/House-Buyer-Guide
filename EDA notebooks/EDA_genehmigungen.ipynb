```python
import requests
import pandas as pd
import json
from cred import credentials
from request_api import get_genesis_api
```

```python
df_genehmigung = pd.read_csv('data/31111-0003 (1).csv', delimiter=';', skiprows=6, skipfooter=4, header=[1,2], encoding = 'latin1')
df_genehmigung
```

```python
df_genehmigung.columns = df_genehmigung.columns.map('_'.join)
df_genehmigung
```

```python
df_genehmigung.columns = df_genehmigung.columns.str.replace(' ', '_')
df_genehmigung
```

```python
df_genehmigung.rename(columns={'Unnamed:_0_level_0_Unnamed:_0_level_1': 'bundesland', 'Unnamed:_1_level_0_Unnamed:_1_level_1': 'building_type', 'Unnamed:_2_level_0_Unnamed:_2_level_1': 'year'}, inplace=True)
df_genehmigung
```

```python
df_genehmigung.drop(columns=['Gebäude_Anzahl', 'Rauminhalt_1000_cbm', 'Baumaßnahmen_Anzahl'], inplace=True)
df_genehmigung
```

```python
df_genehmigung.info()
```

```python
cols = df_genehmigung.columns.drop(['bundesland', 'building_type', 'year'])

df_genehmigung[cols] = df_genehmigung[cols].apply(pd.to_numeric, errors='coerce')
df_genehmigung
```

```python
cols = df_genehmigung.columns.drop(['bundesland', 'building_type', 'year'])
cols
```

```python
df_genehmigung[cols] = df_genehmigung[cols].fillna(-1).astype(int)
df_genehmigung
```
