```python
population_2000_clean = population_2000.query('age_num != "Total"')
```

```python
population_2020_clean = population_2020.query('age_num != "Total"')
population_2020_clean
```

```python
population_2000_clean.rename(columns = {'age_num':'age'}, inplace = True)
population_2000_clean
```

```python
population_2020_clean.rename(columns = {'age_num':'age'}, inplace = True)
population_2020_clean
```

```python
population_2000_clean['age'] = population_2000_clean['age'].map(lambda x: int(x) if x.isnumeric() else 0)
population_2000_clean
```

```python
population_2020_clean['age'] = population_2020_clean['age'].map(lambda x: int(x) if x.isnumeric() else 0)
population_2020_clean
```

```python
def age_to_age_group(age):
    if age < 3:
        return (0,3)
    elif age < 6:
        return (3,6)
    elif age < 10:
        return (6,10)
    elif age < 15:
        return (10,15)
    elif age < 18:
        return (15,18) 
    elif age < 20:
        return (18,20)
    elif age >= 90:
        return (90, np.infty)
    else:
        return (age // 5 * 5, age // 5 * 5 + 5)
    
```

```python
population_2000_clean['age_group'] = population_2000_clean['age'].map(age_to_age_group)
population_2000_clean
```

```python
population_2020_clean['age_group'] = population_2020_clean['age'].map(age_to_age_group)
population_2020_clean
```

```python
population_2000_clean.drop(columns='age', inplace=True)
population_2000_clean
```

```python
population_2020_clean.drop(columns='age', inplace=True)
population_2020_clean
```

```python
grouped_population_2000 = population_2000_clean.groupby('age_group').sum()
grouped_population_2000
```

```python
grouped_population_2020 = population_2000_clean.groupby('age_group').sum()
grouped_population_2020
```

```python
grouped_population_2000['year'] = 2000 
grouped_population_2000
```

```python
grouped_population_2020['year'] = 2000 
grouped_population_2020.info()
```
