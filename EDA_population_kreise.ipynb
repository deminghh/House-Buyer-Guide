```python
df_population_kreise = pd.read_csv('data/12411-0017.csv', delimiter= ';', skiprows= 5, skipfooter=4, encoding= 'latin1', engine='python', converters={'Unnamed: 1': lambda x: str(x)})
df_population_kreise
```

```python
df_population_kreise.rename(columns={'Unnamed: 0': 'date', 'Unnamed: 1': 'key', 'Unnamed: 2': 'kreis'}, inplace=True)
df_population_kreise
```

```python
df_population_kreise.head(30)
```

```python
cols = df_population_kreise.columns.drop(['date', 'key','kreis'])

df_population_kreise[cols] = pd.to_numeric(df_population_kreise[cols].stack(), errors='coerce').unstack()
```

```python
#make NaN values to <NA>
#df_population_kreise[cols] = df_population_kreise[cols].astype('Int64')

#make NaN values to -1

df_population_kreise[cols] = df_population_kreise[cols].fillna(-1).astype(int)
```

```python
#changing date data type to dateime data type
df_population_kreise["date"] = pd.to_datetime(df_population_kreise["date"])
```

```python
df_population_kreise['year'] = pd.DatetimeIndex(df_population_kreise['date']).year
```
